```python
# Définir une fonction pour trouver les doublons en utilisant fuzzy matching
def trouver_doublons(df, colonne, seuil=80):
    valeurs = df[colonne].unique()
    doublons = {}
    
    for valeur in valeurs:
        matches = process.extract(valeur, valeurs, limit=None)
        similars = [match[0] for match in matches if match[1] >= seuil and match[0] != valeur]
        if similars:
            doublons[valeur] = similars
            
    return doublons
```

```python
# Trouver les doublons
doublons = trouver_doublons(df, 'Area')
```

```python
# Définir une fonction pour trouver les doublons en utilisant fuzzy matching
def trouver_doublons(df, colonne, seuil=80, mots_exclus=['island', 'republic', 'united', 'Saint', 'saint']):
    valeurs = df[colonne].unique()
    doublons = {}

    for valeur in valeurs:
        # Exclure les valeurs contenant des mots spécifiques
        if any(mot in valeur.lower() for mot in mots_exclus):
            continue
        
        matches = process.extract(valeur, valeurs, limit=None)
        similars = [match[0] for match in matches if match[1] >= seuil and match[0] != valeur and not any(mot in match[0].lower() for mot in mots_exclus)]
        if similars:
            doublons[valeur] = similars
            
    return doublons

# Trouver les doublons
doublons = trouver_doublons(df, 'Area')
doublons
```
